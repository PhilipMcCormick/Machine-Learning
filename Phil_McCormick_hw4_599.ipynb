```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import cross_validate
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn import datasets
from sklearn.metrics import accuracy_score
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.linear_model import LogisticRegression
```

Build a classifier to predict the Target column using the other 20 features. 

```python
df=pd.read_csv('/content/drive/MyDrive/Data 599/classify_imb.csv')
df.head(1)
```

When you split your data into training and test sets, set random_state = 5. (2 points)

```python
X=df.drop('Target',axis=1)
y=df['Target']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3, random_state=5)
```

Try a random forest (use random_state = 5 and 400 trees), a boosting classifier (use n_estimators = 300, learning rate =0.1, and max_depth =3), and a logistic regression classifier. (9 points: 3 for each classifier)

```python
#random forest
rf = RandomForestClassifier(n_estimators=400,max_features='sqrt',random_state=5).fit(X_train,y_train)
```

```python
accuracy_score(rf.predict(X_test),y_test)
```

```python
pd.crosstab(y_test,rf.predict(X_test),rownames=['True'], colnames=['Predicted'])
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,rf.predict(X_test)))
```

```python
#boosting
bclf = GradientBoostingClassifier(n_estimators=300,max_depth=3,learning_rate=0.1,random_state=5).fit(X_train,y_train)
```

```python
accuracy_score(bclf.predict(X_test),y_test)
```

```python
pd.crosstab(y_test,bclf.predict(X_test),rownames=['True'], colnames=['Predicted'])
```

```python
print(classification_report(y_test,bclf.predict(X_test)))
```

```python
#log reg classifier
```

```python
logreg=LogisticRegression().fit(X_train,y_train)
```

```python
logreg.score(X_test,y_test)
```

```python
print(classification_report(y_test,logreg.predict(X_test)))
```

Which is best? How can you tell? (6 points)

Judging by the f1-score random forest and boosting classifiers did the best at 56% each, with random forest having a slightly stronger accuracy at 77.6%. Therefore, overall, random forest is the best model of the three. The logistic regression di the worst on f1 and accuracy. 

Comment on the precision, recall, and accuracy of your model. (3 points)

Precision, recall, and accuracy are all important metrics depending on what your model is measuring. In regards to our model, since accuracy and f1-score are almost the same in random forest and boosting classification, looking at precision and recall may be most beneficial for which metrics we want our model to capture. If we want to look at how well the model captures observations in Class 1, recall is the important metric to refer to. Boosting outperforms random forest in this area. If we want to know how many predicted 1's were actually 1 then we focus on precision, and random forest beats out boosting in this metric. Random forest beats out boosting slightly in accuracy.
